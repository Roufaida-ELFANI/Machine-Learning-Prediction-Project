# Exploratory Data Analysis

L'objectif de ce projet est de réaliser une analyse exploratoire des données sur un ensemble de données afin d'obtenir une compréhension générale du nombre d'usagers des taxis.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime as dt
import seaborn as sns 
```

```python
df=pd.read_csv('Taxi_Trip_Data.csv')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

Il n'y a pas de données manquantes selon les résultats de la fonction info()

```python
df['tpep_pickup_datetime']=pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime']=pd.to_datetime(df['tpep_dropoff_datetime'])
```

**trip_distance**

```python
sns.boxplot(x=df['trip_distance'], fliersize=1)
```

```python
plt.figure(figsize=(12,4))
ax = sns.histplot(x=df.trip_distance, bins=range(0, 26, 1))
ax.set_xticks(range(0, 26, 1))
ax.set_xticklabels(range(0, 26, 1));
```

La majorité des déplacements sont des trajets de moins de deux miles. Le nombre de déplacements diminue fortement à mesure que la distance parcourue augmente au-delà de deux miles.

**total_amount**

```python
sns.boxplot(x=df['total_amount'], fliersize=1)
```

```python
plt.figure(figsize=(10,4))
ax = sns.histplot(df['total_amount'], bins=range(-10,101,5))
ax.set_xticks(range(-10,101,5))
ax.set_xticklabels(range(-10,101,5));
```

Le coût total de chaque voyage est également réparti vers la droite, la plupart des coûts se situant entre 5 et 15 dollars.

**tip_amount**

```python
plt.title('tip_amount')
sns.boxplot(x=df['tip_amount'], fliersize=1);
```

```python
ax = sns.histplot(df['tip_amount'], bins=range(0,21,1))
ax.set_xticks(range(0,21,2))
ax.set_xticklabels(range(0,21,2));
```

La distribution du tip_amount est orientée vers la droite, la quasi-totalité des pourboires se situant entre 0 et 3 dollars.

**tip_amount par Vendor**

```python
plt.figure(figsize=(12,5))
ax = sns.histplot(data=df, x='tip_amount', bins=range(0, 21, 1), hue='VendorID', multiple='stack', palette='pastel')
ax.set_xticks(range(0, 21, 1))
ax.set_xticklabels(range(0,21,1));
```

La séparation du tip_amount par vendeur révèle qu'il n'y a pas d'aberrations notables dans la distribution des pourboires entre les deux vendeurs. Le deuxième vendeur a une part légèrement plus élevée des courses, et cette proportion se maintient à peu près pour tous les tip_amount.

```python
tips_plus_ten = df[df['tip_amount'] > 10]
plt.figure(figsize=(12,5))
ax = sns.histplot(data=tips_plus_ten, x='tip_amount', bins=range(10,21,1), 
                  hue='VendorID', 
                  multiple='stack',
                  palette='pastel')
ax.set_xticks(range(10,21,1))
ax.set_xticklabels(range(10,21,1));
```

Les proportions sont maintenues même à ces montants de pourboire plus élevés, à l'exception de l'extrémité la plus haute, mais cela ne mérite pas d'être souligné en raison de la faible taille de l'échantillon à ces montants de pourboire.

**Moyenne des pourboires par nombre de passagers**

```python
df['passenger_count'].value_counts()
```

```python
mean_tips_by_passenger_count = df.groupby(['passenger_count'])[['tip_amount']].mean()
mean_tips_by_passenger_count
```

```python
plt.figure(figsize=(10,4))
data = mean_tips_by_passenger_count.tail(-1)
pal = sns.color_palette("Greens_d", len(data))
rank = data['tip_amount'].argsort().argsort()
ax = sns.barplot(x=data.index,
            y=data['tip_amount'],
            palette=np.array(pal[::-1])[rank])
ax.axhline(df['tip_amount'].mean(), ls='--', color='red', label='global mean')
ax.legend();
```

Le montant moyen des pourboires varie très peu en fonction du nombre de passagers. Bien qu'il diminue sensiblement pour les trajets de quatre passagers, on s'attend à une plus grande fluctuation car les trajets de quatre passagers sont les moins nombreux dans l'ensemble de données (à part les trajets de zéro passager).

**Créer les columns 'Month' et 'Day'**

```python
df['month'] = df['tpep_pickup_datetime'].dt.month_name()
df['day'] = df['tpep_pickup_datetime'].dt.day_name()
```

**Plot total ride count par month**

```python
monthly_rides = df['month'].value_counts()
monthly_rides
```

```python
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July',
         'August', 'September', 'October', 'November', 'December']

monthly_rides = monthly_rides.reindex(index=month_order)
monthly_rides
```

```python
plt.figure(figsize=(12,5))
ax = sns.barplot(x=monthly_rides.index, y=monthly_rides)
ax.set_xticklabels(month_order);
```

Les trajets mensuels sont relativement réguliers, avec des baisses notables en juillet, août et septembre, ainsi qu'en février.

**Plot total ride count par day**

```python
daily_rides = df['day'].value_counts()
day_order = ['Saturday', 'Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
daily_rides = daily_rides.reindex(index=day_order)
daily_rides
```

```python
plt.figure(figsize=(10,5))
ax = sns.barplot(x=daily_rides.index, y=daily_rides)
ax.set_xticklabels(day_order)
ax.set_ylabel('Count')
```

C'est du mercredi au samedi que le nombre de trajets quotidiens est le plus élevé, tandis que le dimanche et le lundi sont les moins fréquents.

**Plot total revenue par day**

```python
df['total_amount'] = pd.to_numeric(df['total_amount'], errors='coerce')
total_amount_day = df.groupby('day')['total_amount'].sum()
day_order = ['Saturday', 'Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
total_amount_day = total_amount_day.reindex(index=day_order)
print(total_amount_day)
```

```python
plt.figure(figsize=(10,5))
ax = sns.barplot(x=total_amount_day.index, y=total_amount_day)
ax.set_xticklabels(day_order)
ax.set_ylabel('Revenue (USD)');
```

Le jeudi a eu le revenu le plus élevé de tous les jours, et le dimanche et le lundi ont eu le moins.

**Plot total revenue par month**

```python
total_amount_month = df.groupby('month')[['total_amount']].sum()
total_amount_month = total_amount_month.reindex(index=month_order)
total_amount_month
```

```python
plt.figure(figsize=(12, 5))
sns.barplot(x=total_amount_month.index, y=total_amount_month['total_amount']);
```

les total_amount suivent généralement l'évolution de la fréquentation mensuelle, avec des baisses notables pendant les mois d'été (juillet, août et septembre) et un creux en février.
